```python
from sklearn.decomposition import PCA
import pandas as pd
import plotly.express as px

digits = datasets.load_digits()
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
target = digits.target

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data)

df_pca = pd.DataFrame({'target': map(lambda x : str(x),target), 'x': data_pca[:, 0], 'y': data_pca[:, 1]})
fig = px.scatter(df_pca, x="x", y="y", color="target", title ="PCA projected data", labels={'x': 'PCA 1', 'y': 'PCA 2', 'target': 'Digit'})
fig.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=2)
data_lda = lda.fit_transform(data, target)

df_lda = pd.DataFrame({'target': map(lambda x : str(x),target), 'x': data_lda[:, 0], 'y': data_lda[:, 1]})
fig = px.scatter(df_lda, x="x", y="y", color="target", title ="LDA projected data", labels={'x': 'LDA 1', 'y': 'LDA 2', 'target': 'Digit'})
fig.show()
```

```python
from sklearn.manifold import MDS
mds = MDS(n_components=2)
data_mds = mds.fit_transform(data)

df_mds = pd.DataFrame({'target': map(lambda x : str(x),target), 'x': data_mds[:, 0], 'y': data_mds[:, 1]})
fig = px.scatter(df_mds, x="x", y="y", color="target", title ="MDS projected data", labels={'x': 'MDS 1', 'y': 'MDS 2', 'target': 'Digit'})
fig.show()
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=2)
data_svd = svd.fit_transform(data)

df_svd = pd.DataFrame({'target': map(lambda x : str(x),target), 'x': data_svd[:, 0], 'y': data_svd[:, 1]})
fig = px.scatter(df_svd, x="x", y="y", color="target")
fig.show()
```

```python
from sklearn.manifold import Isomap
isomap = Isomap(n_components=2)
data_isomap = isomap.fit_transform(data)

df_isomap = pd.DataFrame({'target': map(lambda x : str(x),target), 'x': data_isomap[:, 0], 'y': data_isomap[:, 1]})
fig = px.scatter(df_isomap, x="x", y="y", color="target", title ="Isomap projected data", labels={'x': 'Isomap 1', 'y': 'Isomap 2', 'target': 'Digit'})
fig.show()
```

```python
from sklearn.decomposition import FastICA
ica = FastICA(n_components=2)
data_ica = ica.fit_transform(data)

df_ica = pd.DataFrame({'target': map(lambda x : str(x),target), 'x': data_ica[:, 0], 'y': data_ica[:, 1]})
fig = px.scatter(df_ica, x="x", y="y", color="target", title ="ICA projected data", labels={'x': 'ICA 1', 'y': 'ICA 2', 'target': 'Digit'})
fig.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)
data_tsne = tsne.fit_transform(data)

data_tsne = pd.DataFrame({'target': map(lambda x : str(x),target), 'x': data_tsne[:, 0], 'y': data_tsne[:, 1]})
fig = px.scatter(data_tsne, x="x", y="y", color="target", title ="TSNE projected data", labels={'x': 'TSNE 1', 'y': 'TSNE 2', 'target': 'Digit'})
fig.show()
```
